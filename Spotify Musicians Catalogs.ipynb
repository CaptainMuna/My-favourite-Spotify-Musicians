# Artist

```python
def get_artist(names):
    results = [sp.search(q=x, type ='artist', market = 'US', limit =1) for x in names]
    return results
```

```python
# Getting artist Names from the JSON 
names = ['Rema', 'Fireboy DML','Ayra Starr','Burna Boy','Davido', 'Wizkid', 'Asake', 'Olamide', 'Ckay','Tems', 'Oxlade',
        'Joeboy', 'Wurld', 'Zinoleesky', 'Falz', 'Zlatan', 'Phyno', 'Omah Lay', 'Lojay', 'Drake']
results = get_artist(names)
results
```

```python
# Populating the artist table with data
def artist_info(result):
    artist = {}
    artist['artist_name'] = [result[i]['artists']['items'][0]['name'] for i in range(len(result))]
    artist['id'] = [result[i]['artists']['items'][0]['id'] for i in range(len(result))]
    artist['type'] = [result[i]['artists']['items'][0]['type'] for i in range(len(result))]
    artist['external_url'] = [result[i]['artists']['items'][0]['external_urls']['spotify'] for i in range(len(result))]
    artist['followers'] = [result[i]['artists']['items'][0]['followers']['total'] for i in range(len(result))]
    artist['genres'] = [result[i]['artists']['items'][0]['genres'][0] for i in range(len(result))]
    artist['popularity'] = [result[i]['artists']['items'][0]['popularity'] for i in range(len(result))]
    artist['artist_uri'] = [result[i]['artists']['items'][0]['uri'] for i in range(len(result))]
    return artist
```
